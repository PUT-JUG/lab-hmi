```python
import numpy as np
def dft(x):
    x = np.asarray(x, dtype=float)
    N = x.shape[0]
    n = np.arange(N)
    k = n.reshape((N, 1))
    M = np.exp(-2j * np.pi * k * n / N)
    return np.dot(M, x)
```

```python
import pylab as py
import numpy as np
from numpy.fft import rfft, rfftfreq

def sin(f = 1, T = 1, Fs = 128, phi =0 ):
    dt = 1.0/Fs
    t = np.linspace(0.0, T, int(T*Fs), endpoint=False)
    y = np.sin(2.0*np.pi*f*t)
    return (y,t)
```

```python
from scipy.fft import fft, fftfreq
Fs =100
T =1

(y,t) = sin(f = 10.0, T=T, Fs=Fs)

N=int(Fs*T)

yf = fft(y)
xf = fftfreq(N, 1/Fs)
import matplotlib.pyplot as plt
plt.stem(xf, Fs/N * np.abs(yf))
plt.grid()
plt.show()
```
